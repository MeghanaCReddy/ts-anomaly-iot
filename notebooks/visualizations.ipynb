```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_features = pd.read_csv(r"C:\Users\Meghana C Reddy\Desktop\ts_work\ts-anomaly-iot\data\processed\features.csv")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 5))
plt.plot(df_features.index[:1000], df_features['ch_1'][:1000], label='ch_1')
plt.plot(df_features.index[:1000], df_features['ch_2'][:1000], label='ch_2')
plt.title("Sample Raw Sensor Signals (First 1000 readings)")
plt.xlabel("Time Index")
plt.ylabel("Sensor Value")
plt.legend()
plt.tight_layout()
plt.savefig("reports/figures/raw_sensor_signal.png")
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.hist(df_features['ch_1'], bins=50, alpha=0.7, color='blue', label='ch_1')
plt.hist(df_features['ch_2'], bins=50, alpha=0.5, color='green', label='ch_2')
plt.title("Feature Distribution: Rolling Mean")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.tight_layout()
plt.savefig("reports/figures/feature_distribution.png")
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.plot(df_features.index[:1000], df_features['ch_1'][:1000], label='ch_1 signal')
plt.scatter(df_features.index[:1000][df_features['iso_anomaly'][:1000]==1],
            df_features['ch_1'][:1000][df_features['iso_anomaly'][:1000]==1],
            color='red', label='Isolation Forest Anomalies', s=15)
plt.title("Isolation Forest: Anomaly Detection on ch_1")
plt.xlabel("Time Index")
plt.ylabel("Sensor Value")
plt.legend()
plt.tight_layout()
plt.savefig("reports/figures/iso_forest_anomalies.png")
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.plot(df_features.index[:1000], df_features['ch_1'][:1000], label='ch_1 signal')
plt.scatter(df_features.index[:1000][df_features['ae_anomaly'][:1000]==1],
            df_features['ch_1'][:1000][df_features['ae_anomaly'][:1000]==1],
            color='green', label='Autoencoder Anomalies', s=15)
plt.title("Autoencoder: Anomaly Detection on ch_1")
plt.xlabel("Time Index")
plt.ylabel("Sensor Value")
plt.legend()
plt.tight_layout()
plt.savefig("reports/figures/autoencoder_anomalies.png")
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.plot(df_features.index[:1000], df_features['ch_1'][:1000], label='ch_1 signal')
plt.scatter(df_features.index[:1000][df_features['iso_anomaly'][:1000]==1],
            df_features['ch_1'][:1000][df_features['iso_anomaly'][:1000]==1],
            color='red', label='Isolation Forest', s=15)
plt.scatter(df_features.index[:1000][df_features['ae_anomaly'][:1000]==1],
            df_features['ch_1'][:1000][df_features['ae_anomaly'][:1000]==1],
            color='green', label='Autoencoder', s=15, marker='x')
plt.title("Anomaly Detection Comparison on ch_1")
plt.xlabel("Time Index")
plt.ylabel("Sensor Value")
plt.legend()
plt.tight_layout()
plt.savefig("reports/figures/anomaly_comparison.png")
plt.show()
```
